```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast
import glob

# Set Seaborn style
sns.set(style="whitegrid")

# Function to process the server nodes data
def process_nodes(file_path):
    df = pd.read_csv(file_path)
    df['dcount_cloud_RoleInstance'] = df['dcount_cloud_RoleInstance'].apply(ast.literal_eval)
    df['timestamp'] = df['timestamp'].apply(ast.literal_eval)

    flattened_data = []
    for role_instance, timestamps in zip(df['dcount_cloud_RoleInstance'], df['timestamp']):
        for count, time in zip(role_instance, timestamps):
            flattened_data.append((int(count), pd.to_datetime(time)))

    flattened_df = pd.DataFrame(flattened_data, columns=['dcount_cloud_RoleInstance', 'timestamp'])
    flattened_df['normalized_timestamp'] = (flattened_df['timestamp'] - flattened_df['timestamp'].min()).dt.total_seconds()
    max_time = flattened_df['normalized_timestamp'].max()
    flattened_df['rescaled_timestamp'] = flattened_df['normalized_timestamp'] / max_time

    return flattened_df[['rescaled_timestamp', 'dcount_cloud_RoleInstance']]

# Function to process CPU and Memory data
def process_cpu_memory(file_path, column_name, is_percentage=False):
    df = pd.read_csv(file_path)
    
    # If the data is in percentage, remove the '%' sign and convert to float
    if is_percentage:
        df[column_name] = df[column_name].str.replace('%', '').astype(float)
    else:
        df[column_name] = df[column_name].str.replace(' MB', '').astype(float)
    
    df['Time'] = pd.to_datetime(df['Time'])
    df['normalized_timestamp'] = (df['Time'] - df['Time'].min()).dt.total_seconds()
    max_time = df['normalized_timestamp'].max()
    df['rescaled_timestamp'] = df['normalized_timestamp'] / max_time
    return df[['rescaled_timestamp', column_name, 'Time']]

# ------------------- Graph 1: Node Scaling vs. Resource Utilization Over Time -------------------

# Prepare the plot
fig, ax1 = plt.subplots(figsize=(14, 8))

# Process server nodes data
nodes_df = process_nodes('azure/fhir/exp-4-1-Autoscale.csv')
ax1.plot(nodes_df['rescaled_timestamp'], nodes_df['dcount_cloud_RoleInstance'], label='Server Nodes', color='blue', linewidth=2)
ax1.set_xlabel('Rescaled Time', fontsize=14)
ax1.set_ylabel('Number of Server Nodes', fontsize=14, color='blue')

# Process CPU usage data
cpu_df = process_cpu_memory('azure/fhir/cpu/exp-4-1-cpu.csv', 'Process CPU - healthcareorchestration', is_percentage=True)
ax2 = ax1.twinx()
ax2.plot(cpu_df['rescaled_timestamp'], cpu_df['Process CPU - healthcareorchestration'], label='CPU Usage', color='red', linewidth=2, linestyle='--')
ax2.set_ylabel('CPU Usage (%)', fontsize=14, color='red')

# Process Memory usage data
memory_df = process_cpu_memory('azure/fhir/memory/exp-4-1-memory.csv', 'Process private bytes - healthcareorchestration')
ax2.plot(memory_df['rescaled_timestamp'], memory_df['Process private bytes - healthcareorchestration'], label='Memory Usage', color='green', linewidth=2, linestyle='-.')
ax2.set_ylabel('Memory Usage (MB)', fontsize=14, color='green')

# Add legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.title('Node Scaling vs. Resource Utilization Over Time', fontsize=18, weight='bold')
plt.grid(False)  # Disable the grid
plt.tight_layout()
plt.savefig("node_scaling_vs_resource_utilization.png", dpi=300)
plt.show()

# ------------------- Graph 2: Correlation Between Nodes, CPU, and Memory Usage -------------------

# Prepare the plot
fig, ax = plt.subplots(figsize=(14, 8))

# Scatter plot for CPU usage vs. Nodes
sns.scatterplot(x=nodes_df['dcount_cloud_RoleInstance'], y=cpu_df['Process CPU - healthcareorchestration'], ax=ax, label='CPU Usage vs. Nodes', color='red')

# Scatter plot for Memory usage vs. Nodes
sns.scatterplot(x=nodes_df['dcount_cloud_RoleInstance'], y=memory_df['Process private bytes - healthcareorchestration'], ax=ax, label='Memory Usage vs. Nodes', color='green')

ax.set_xlabel('Number of Server Nodes', fontsize=14)
ax.set_ylabel('Resource Usage', fontsize=14)
plt.title('Correlation Between Nodes, CPU, and Memory Usage', fontsize=18, weight='bold')
plt.grid(False)  # Disable the grid
plt.legend(loc='upper left')
plt.tight_layout()
plt.savefig("correlation_between_nodes_cpu_memory.png", dpi=300)
plt.show()

# ------------------- Graph 3: Temporal Resource Utilization -------------------

# Prepare the plot
fig, ax = plt.subplots(figsize=(14, 8))

# Plot CPU and Memory usage over Time
sns.lineplot(x=cpu_df['Time'], y=cpu_df['Process CPU - healthcareorchestration'], ax=ax, label='CPU Usage', color='red')
sns.lineplot(x=memory_df['Time'], y=memory_df['Process private bytes - healthcareorchestration'], ax=ax, label='Memory Usage', color='green')

ax.set_xlabel('Time', fontsize=14)
ax.set_ylabel('Resource Usage', fontsize=14)
plt.title('Temporal Resource Utilization', fontsize=18, weight='bold')
plt.grid(False)  # Disable the grid
plt.legend(loc='upper right')
plt.tight_layout()
plt.savefig("temporal_resource_utilization.png", dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for Healthcare Orchestration Pipeline
azure_healthcare_executions = [4924, 3824, 6235, 3699, 6261]
knative_healthcare_executions = [2719, 2843, 2638, 2483, 2499]
executions_labels = ['Execution 1', 'Execution 2', 'Execution 3', 'Execution 4', 'Execution 5']

azure_healthcare_avg = np.mean(azure_healthcare_executions)
knative_healthcare_avg = np.mean(knative_healthcare_executions)

# Data for Data Processing Pipeline
azure_data_processing_executions = [8277, 8456, 5971, 15109, 9536]
knative_data_processing_executions = [2512, 3073, 2836, 2877, 2538]

azure_data_processing_avg = np.mean(azure_data_processing_executions)
knative_data_processing_avg = np.mean(knative_data_processing_executions)

# Create the plots
plt.figure(figsize=(14, 10))

# Plot for Healthcare Orchestration Pipeline
plt.subplot(2, 1, 1)
plt.plot(executions_labels, azure_healthcare_executions, marker='o', linestyle='-', color='blue', label='Azure Functions')
plt.plot(executions_labels, knative_healthcare_executions, marker='o', linestyle='-', color='green', label='Knative')
plt.axhline(y=azure_healthcare_avg, color='blue', linestyle='--', label=f'Azure Avg ({azure_healthcare_avg:.2f} ms)')
plt.axhline(y=knative_healthcare_avg, color='green', linestyle='--', label=f'Knative Avg ({knative_healthcare_avg:.2f} ms)')
plt.title('Healthcare Orchestration Pipeline: Azure Functions vs Knative')
plt.xlabel('Execution')
plt.ylabel('Load Time (ms)')
plt.ylim(0, max(max(azure_healthcare_executions), max(knative_healthcare_executions)) + 2000)
plt.legend()
plt.grid(True)

# Plot for Data Processing Pipeline
plt.subplot(2, 1, 2)
plt.plot(executions_labels, azure_data_processing_executions, marker='o', linestyle='-', color='blue', label='Azure Functions')
plt.plot(executions_labels, knative_data_processing_executions, marker='o', linestyle='-', color='green', label='Knative')
plt.axhline(y=azure_data_processing_avg, color='blue', linestyle='--', label=f'Azure Avg ({azure_data_processing_avg:.2f} ms)')
plt.axhline(y=knative_data_processing_avg, color='green', linestyle='--', label=f'Knative Avg ({knative_data_processing_avg:.2f} ms)')
plt.title('Data Processing Pipeline: Azure Functions vs Knative')
plt.xlabel('Execution')
plt.ylabel('Load Time (ms)')
plt.ylim(0, max(max(azure_data_processing_executions), max(knative_data_processing_executions)) + 3000)
plt.legend()
plt.grid(True)

# Display the plots
plt.tight_layout()
plt.savefig("COLD.png", dpi=300)
plt.show()
```
